```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import make_axes_locatable
from READ_ART_bigendian import read_ART
import scipy
from scipy import stats
import glob
import cv2
```

```python
# test = "test_10_times_less_particles"
test = "test_high_resolution_wsplit=0.5_wjoin=0.25"
```

```python
snapshots = glob.glob("/home/hpc/klumpke/ojimenez/Escritorio/phD/2021/OCRE_simulations/ART_pure_Nbody-master/RUN/snapshots/"+str(test)+"/PMcrs0a*")
snapshots = [i[-10:-4] for i in snapshots]
snapshots
```

```python
# a = '0.6850'
MW = 'yes'
zoom = 'no'
save_csv = 'no'

for a in snapshots:
    print ('a = ',a)
    if test == "test_10_times_less_particles":
        path = '/home/hpc/klumpke/ojimenez/Escritorio/phD/2021/OCRE_simulations/ART_pure_Nbody-master/RODIN/'+test+'/LMC+SMC+MW/'
    elif test == "test_high_resolution_wsplit=0.5_wjoin=0.25":
        path = '/home/hpc/klumpke/ojimenez/Escritorio/phD/2021/OCRE_simulations/ART_pure_Nbody-master/RODIN/LMC_relaxed+SMC+MW/'
    path2 = '/home/hpc/klumpke/ojimenez/Escritorio/phD/2021/OCRE_simulations/ART_pure_Nbody-master/RUN/snapshots/'+test+'/'
    filename = 'PMcrs0a'+a+'.DAT'
    f_LMC = open(path+"Rodin2.out_LMC", "r")
    f_SMC = open(path+"Rodin2.out_SMC", "r")
#     f_MW = open(path+"Rodin2.out_MW", "r")
    nstars_LMC = int(f_LMC.read().splitlines()[16][-9:])
    nstars_SMC = int(f_SMC.read().splitlines()[-1][-9:])
#     nstars_MW = int(f_MW.read().splitlines()[-1][-9:])
#     nstars_MW = 1
    nstars = nstars_LMC + nstars_SMC #+ nstars_MW
    print ('filename: '+str(filename))
    print ('Galaxy: LMC relaxed + SMC + MW (snapshot)')
    print ('')
    print ('Nstars (LMC) =    '+str(nstars_LMC))
    print ('Nstars (SMC) =     '+str(nstars_SMC))
#     print ('Nstars (MW) =           '+str(nstars_MW))
    print ('Nstars (total) =  '+str(nstars))
    print ('-----------------------------------------------------------------------------------------')
    mass,x,y,z,vx,vy,vz,Id=read_ART(path2,filename,nstars)
    
    df = pd.DataFrame()
    df['Id'] = Id[0]
    df['x'], df['y'], df['z'] = x[0], y[0], z[0]
    df['vx'], df['vy'], df['vz'] = vx[0], vy[0], vz[0]
    df['mass'] = mass[0]
    df['galaxy'] = 'LMC'
    df['specie'] = 'Star'
    df.loc[0:nstars_SMC, 'galaxy'] = 'SMC'

    for i in range(1,len(x)):
        df_aux = pd.DataFrame()
        df_aux['Id'] = Id[i]
        df_aux['x'], df_aux['y'], df_aux['z'] = x[i], y[i], z[i]
        df_aux['vx'], df_aux['vy'], df_aux['vz'] = vx[i], vy[i], vz[i]
        df_aux['mass'] = mass[i]
        df_aux['galaxy'] = 'LMC'
        df_aux['specie'] = 'DM'+str(i)
        df = pd.concat([df,df_aux], ignore_index=True)

    if MW == 'yes':
        df.loc[(df['galaxy']=='LMC') & (df['specie']=='DM'+str(len(x)-1)), ['galaxy','specie']] = ['MW', 'point-like']

    print ('Unique values from Specie and Mass Columns:')
    print ((df['specie']+' '+df['mass'].astype(str)).unique())

    mask_LMC_star = (df['galaxy'] == 'LMC') & (df['specie'] == 'Star')
    mask_LMC_DM1 = (df['galaxy'] == 'LMC') & (df['specie'] == 'DM1')
    mask_LMC_DM2 = (df['galaxy'] == 'LMC') & (df['specie'] == 'DM2')
    mask_LMC_DM3 = (df['galaxy'] == 'LMC') & (df['specie'] == 'DM3')
    mask_SMC_star = (df['galaxy'] == 'SMC') & (df['specie'] == 'Star')
    mask_MW = (df['galaxy'] == 'MW') & (df['specie'] == 'point-like')

    mask_LMC_star_SMC_star = mask_LMC_star | mask_SMC_star
    mask_LMC_DM = mask_LMC_DM1 | mask_LMC_DM2 | mask_LMC_DM3

    masks = [mask_LMC_star_SMC_star, mask_LMC_DM, mask_LMC_DM1, mask_LMC_DM2, mask_LMC_DM3]
    filetext = ['LMC_stars_SMC', 'LMC_DM', 'LMC_DM1', 'LMC_DM2', 'LMC_DM3']
    title = ['LMC Stars + SMC', 'LMC DM', 'LMC DM1', 'LMC DM2', 'LMC DM3']
    
    mask_LMC = df['galaxy']=='LMC'
    x_LMC_median = np.median(df[mask_LMC]['x'])
    y_LMC_median = np.median(df[mask_LMC]['y'])
    z_LMC_median = np.median(df[mask_LMC]['z'])

    df['x'] = df['x'] - x_LMC_median
    df['y'] = df['y'] - y_LMC_median
    df['z'] = df['z'] - z_LMC_median
    
    print ('LMC stellar mass:   '+ "{:e}".format(np.sum(df[mask_LMC_star]['mass']))+ ' Msun')
    print ('LMC DM mass:        '+ "{:e}".format(np.sum(df[mask_LMC_DM]['mass']))+ ' Msun')
    print ('LMC total mass:     '+ "{:e}".format(np.sum(df[df['galaxy']=='LMC']['mass']))+ ' Msun')
    print ('------------------------------------------')
    print ('SMC stellar mass:   '+ "{:e}".format(np.sum(df[mask_SMC_star]['mass']))+ ' Msun')
    print ('------------------------------------------')
    print ('MW mass:            '+ "{:e}".format(np.sum(df[mask_MW]['mass']))+ ' Msun')
    
    nbins = 300

    if zoom == 'no':
        xmin, xmax = -200, 200
        ymin, ymax = -200, 200
        vmax = [2e4, 7e2]
    elif zoom == 'yes':
        xmin, xmax = -20, 20
        ymin, ymax = -20, 20
        vmax = [0.8e2, 7e2]

#     for i in range(2):
    for i in range(1):
        fig = plt.figure(figsize=(12,8))
        ax = fig.add_subplot(121)
        im = plt.hist2d(df[masks[i]]['x'], df[masks[i]]['y'], bins=(nbins, nbins), range=[[xmin, xmax],[ymin, ymax]], norm=mpl.colors.LogNorm(vmin=1, vmax=vmax[i]), cmap=plt.cm.gnuplot2_r)
        plt.scatter(df[mask_MW]['x'], df[mask_MW]['y'], s=20, c='k')
        plt.xlabel(r"$x$ [kpc]")
        plt.ylabel(r"$y$ [kpc]")
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        plt.colorbar(im[3], cax=cax)
        ax.set_title(r'Density plot ['+str(title[i])+'] ('+str(len(df[masks[i]]))+')')
        if zoom == 'no':
            ax.text(55,-170, 'a = '+a, fontsize=18, bbox=dict(facecolor='white', edgecolor='black'))
        if zoom == 'yes':
            ax.text(5.5,-17, 'a = '+a, fontsize=18, bbox=dict(facecolor='white', edgecolor='black'))
        ax.set_xlim(xmin, xmax)
        ax.set_ylim(ymin, ymax)
        ax.set_aspect('equal', adjustable='box')

        ax2 = fig.add_subplot(122)
        im2 = plt.hist2d(df[masks[i]]['x'], df[masks[i]]['z'], bins=(nbins, nbins), range=[[xmin, xmax],[ymin, ymax]], norm=mpl.colors.LogNorm(vmin=1, vmax=vmax[i]), cmap=plt.cm.gnuplot2_r)
        plt.scatter(df[mask_MW]['x'], df[mask_MW]['z'], s=20, c='k')
        plt.xlabel(r"$x$ [kpc]")
        plt.ylabel(r"$z$ [kpc]")
        divider2 = make_axes_locatable(ax2)
        cax2 = divider2.append_axes("right", size="5%", pad=0.05)
        plt.colorbar(im2[3], cax=cax2)
        ax2.set_title(r'Density plot ['+str(title[i])+'] ('+str(len(df[masks[i]]))+')')
        if zoom == 'no':
            ax2.text(55,-170, 'a = '+a, fontsize=18, bbox=dict(facecolor='white', edgecolor='black'))
        if zoom == 'yes':
            ax2.text(5.5,-17, 'a = '+a, fontsize=18, bbox=dict(facecolor='white', edgecolor='black'))
        ax2.set_xlim(xmin, xmax)
        ax2.set_ylim(ymin, ymax)
        ax2.set_aspect('equal', adjustable='box')

        fig.tight_layout()
        if zoom == 'no':
            fig.savefig(path2+'plots/'+"{:02d}".format(i)+'.PMcrs0a'+a+'_'+filetext[i]+'.png', bbox_inches='tight', facecolor='w')
        elif zoom == 'yes':
            fig.savefig(path2+'plots/'+"{:02d}".format(i)+'.PMcrs0a'+a+'_'+filetext[i]+'_zoom.png', bbox_inches='tight', facecolor='w')
        plt.show()
        
        print ('---------------------------------------------------------------------')
       
    if save_csv == 'yes':
        print ('Save the dataframe into .csv...')
        df[masks[0]].to_csv(path2+'csv/PMcrs0a'+str(a)+'_'+filetext[0]+'.csv')
```
